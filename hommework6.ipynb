```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import torch.optim as optim
import torch.nn as nn
import pandas as pd
import torch

from google.colab import drive
drive.mount('/content/drive')

# Load your dataset
file_path = '/content/drive/My Drive/Colab Notebooks/Housing.csv'
housing = pd.DataFrame(pd.read_csv(file_path))

varlist = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, 'no': 0})

# Applying binary mapping to specified variables
housing[varlist] = housing[varlist].apply(binary_map)

# Dropping the 'furnishingstatus' column
housing = housing.drop('furnishingstatus', axis=1)

# Separating target variable ('price') and features
y = housing.pop('price').values.reshape(-1, 1)
features = housing.columns.tolist()
x = housing.values

x_n = torch.tensor(StandardScaler().fit_transform(x), dtype=torch.float32)
y_n = torch.tensor(StandardScaler().fit_transform(y), dtype=torch.float32)

x_train, x_valid, y_train, y_valid = train_test_split(x_n, y_n, test_size=0.8, random_state=40)

# Define the neural network model with increased complexity
seq_model = nn.Sequential(
    nn.Linear(11, 32),
    nn.Tanh(),
    nn.Linear(32, 64),
    nn.Tanh(),
    nn.Linear(64, 16),
    nn.Tanh(),
    nn.Linear(16, 1))

# Define the optimizer
optimizer = optim.SGD(seq_model.parameters(), lr=1e-3)

# Define the training function
def training(epochs, optimizer, model, loss_fn, x_train, x_valid, y_train, y_valid):
    for epoch in range(1, epochs + 1):
        model.train()
        y_pred = model(x_train)
        loss = loss_fn(y_pred, y_train)
        y_val_pred = model(x_valid)
        val_loss = loss_fn(y_val_pred, y_valid)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if epoch % 500 == 0:
            print(f'Epoch [{epoch}], Training Loss: {loss.item():.4f}, Validation Loss: {val_loss.item():.4f}')

# Training the model
training(epochs=5000, optimizer=optimizer, model=seq_model, loss_fn=nn.MSELoss(), x_train=x_train, x_valid=x_valid, y_train=y_train, y_valid=y_valid)
```
